```python
import pandas as pd
import numpy as np

environment_file = "/Users/fordfishman/GitHub/envs110/assignment-4/data/Arrigetch_Peaks_Environmental_Data.csv"
species_file = "/Users/fordfishman/GitHub/envs110/assignment-4/data/Arrigetch_Peaks_Species_Data.csv"

env = pd.read_csv(environment_file)
species = pd.read_csv(species_file)
# loaded data set
```

```python
# replaces values
species = species.replace(-9999, np.NaN)
env = env.replace(-9999, np.NaN)
```

```python
num_cols = ['ELEVATION', 'ASPECT','SLOPE','COVER_LITTER_LAYER', 'COVER_OPEN_WATER','COVER_ROCK','COVER_CRUST', 'COVER_BARE_SOIL']

env[num_cols].describe()
```

```python
merged_data = pd.merge(env, species, left_on='TURBOVEG_PLOT_NUMBER', right_on='TURBOVEG_PLOT_NUMBER')

sum(merged_data.isnull().any(axis=1))
```

```python

species_counts = {}

for col in species.columns: 
    
    if col != 'TURBOVEG_PLOT_NUMBER': # for all species columns
        
        species_counts[col] = species[col][species[col]>0].shape[0] # gets back how many samples a species was in
  
max_species = ''
m = 0  
        
for species_name, count in species_counts.items(): # find the maximum amount of times a species was found
    
    if count > m:
        m = count
        max_species = species_name

print(max_species)
```
